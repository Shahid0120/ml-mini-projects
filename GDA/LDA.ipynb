# Testing LDA for two feature values - petal width and sepal width 

```python
# Visualising
data = iris.data
labels = iris.target
label_names = iris.target_names

# Extract sepal width and petal width
sepal_width = data[:, 1] 
petal_width = data[:, 3] 

# Plot the data
plt.figure(figsize=(8, 6))
for i in range(len(label_names)):
    plt.scatter(sepal_width[labels == i], petal_width[labels == i], label=label_names[i])

plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Plants: Sepal Width vs Petal Width')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Load the Iris dataset
iris = sns.load_dataset('iris')

# Filter out the 'versicolor' species
iris_filtered = iris[iris['species'] != 'versicolor']

# Create a FacetGrid for each species
g = sns.FacetGrid(iris_filtered, hue='species', height=8)

# Plot the data with kdeplot to show Gaussian contours
g.map(sns.kdeplot, 'sepal_length', 'sepal_width', fill=True, alpha=0.5)

# Add scatter plot on top
g.map(sns.scatterplot, 'sepal_length', 'sepal_width')

# Add a legend
g.add_legend()

# Set axis labels
g.set_axis_labels('Sepal Length', 'Sepal Width')

# Show the plot
plt.show()
```

Clearly is it linearly seperable, but from visual inspection, both seem to have different covariance matrix's

# Naive Bayes

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Plot Gaussian distribution for Sepal Length for each class
plt.figure(figsize=(14, 6))

# Sepal Length
plt.subplot(1, 2, 1)
sns.histplot(data=iris, x='sepal_length', hue='species', kde=True, element='step')
plt.title('Gaussian Distribution of Sepal Length for Each Species')
plt.xlabel('Sepal Length')
plt.ylabel('Density')

# Sepal Width
plt.subplot(1, 2, 2)
sns.histplot(data=iris, x='sepal_width', hue='species', kde=True, element='step')
plt.title('Gaussian Distribution of Sepal Width for Each Species')
plt.xlabel('Sepal Width')
plt.ylabel('Density')

# Show the plots
plt.tight_layout()
plt.show()
```
