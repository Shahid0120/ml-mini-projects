# Understanding probability integral transform

Suppose i conduct a random experiment for and for inference i beleive the data comes from a paremtric method, which i assume normally distribtued X ~ Exp(1).
According to the Probaiblity Integral Transformation if Y = F_x(x) then Y~U[0,1]. 
So 
Fx(x) = 1 - exp(\lambda * x)
Y = Fx(x) 

then 

y = 1 - exp(\lambda * x) <=> x = -log(1 - y).


So me generating Y~U(0,1) then applying transformation -log(1-Y) should then be all generating X ~ Exp(1).

```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy
import scipy.stats as stats
```

```python
# generating and plot some random points usi
try:
    scale = 1
    print("Generating 1000 normal samples....")
    x = np.arange(0.1, 50, 0.1)
    print("Generated 1000 normal samples....")
    y = scipy.stats.expon.cdf(x, scale=scale)
    
    plt.figure()
    plt.title("Plot of generated function from X ~ Exp(1)")
    plt.plot(x, y)
    plt.show()
    
    print("Generating from Z~U[0,1]....")
    Z = np.random.uniform(0, 1, np.shape(x)[0])
    print("Generated points from Y~U[0,1]...")
    
    # applying x = -log(1-y)
    print("Applying inverse transformation")
    generated_exp = -np.log(1 - Z)
    print("Applied inverse transformation")
    
    plt.figure()
    plt.title("Plot of generated function from Y~U[0,1] then transformation X = -log(1 - Y)")
    plt.plot(np.sort(generated_exp), np.arange(len(generated_exp)) / len(generated_exp))
    plt.show()

    
    plt.figure()
    plt.title("Empirical vs. Theoretical CDF")
    # Empirical CDF from samples
    plt.plot(np.sort(generated_exp), np.arange(1, len(generated_exp) + 1) / len(generated_exp), label="Empirical CDF")
    # Theoretical CDF of the exponential distribution
    plt.plot(x, scipy.stats.expon.cdf(x, scale=scale), label="Theoretical CDF", linestyle="dashed")
    plt.legend()
    plt.show()

    err = np.linalg.norm(np.sort(y) - np.sort(Z), 1)
    print("Error", err)

    
    
except:
    print("something went wrong")
```
