```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate a dataset with 8 points (2D points for visualization purposes)
np.random.seed(42)
points = np.random.rand(8, 2)

# Perform hierarchical clustering
Z = linkage(points, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z, labels=np.arange(1, 9))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
# divisive clustering
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from scipy.spatial import distance_matrix
from scipy.sparse.csgraph import minimum_spanning_tree

# Step 1: Generate random points
np.random.seed(0)
points = np.random.rand(5, 2)

# Step 2: Calculate the Euclidean distance matrix
dist_matrix = distance_matrix(points, points)

# Step 3: Construct the Minimum Spanning Tree (MST)
mst = minimum_spanning_tree(dist_matrix).toarray()

# Step 4: Plot the points and the MST
fig, ax = plt.subplots()
ax.scatter(points[:, 0], points[:, 1], color='blue')

# Annotate points
for i, (x, y) in enumerate(points):
    ax.annotate(f'P{i+1}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

# Draw the MST edges
for i in range(len(points)):
    for j in range(len(points)):
        if mst[i, j] > 0:
            ax.plot([points[i, 0], points[j, 0]], [points[i, 1], points[j, 1]], 'k-', lw=1.5)

# Add ellipse to make cluster
ellipse1 = Ellipse((0.5, 0.5), width=1, height=1, angle=0, color='red', fill=False)
ax.add_patch(ellipse1)
plt.title("Minimum Spanning Tree of Average income on Age")
plt.xlabel("Average income (1,000s)")
plt.ylabel("Age (years)")
plt.grid(True)
plt.show()
```

```python
# divisive clustering
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from scipy.spatial import distance_matrix
from scipy.sparse.csgraph import minimum_spanning_tree

# Step 1: Generate random points
np.random.seed(0)
points = np.random.rand(5, 2)

# Step 2: Calculate the Euclidean distance matrix
dist_matrix = distance_matrix(points, points)

# Step 3: Construct the Minimum Spanning Tree (MST)
mst = minimum_spanning_tree(dist_matrix).toarray()

# Step 4: Plot the points and the MST
fig, ax = plt.subplots()
ax.scatter(points[:, 0], points[:, 1], color='blue')

# Annotate points
for i, (x, y) in enumerate(points):
    ax.annotate(f'P{i+1}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

# Draw the MST edges
for i in range(len(points)):
    for j in range(len(points)):
        if mst[i, j] > 0:
            ax.plot([points[i, 0], points[j, 0]], [points[i, 1], points[j, 1]], 'k-', lw=1.5)

# Add ellipse to make cluster
ellipse1 = Ellipse((0.5, 0.7), width=0.5, height=0.5, angle=0, color='red', fill=False)
ax.add_patch(ellipse1)

ellipse2 = Ellipse((0.95, 0.4), width=0.2, height=0.2, angle=0, color='purple', fill=False)
ax.add_patch(ellipse2)

plt.title("Minimum Spanning Tree of Average income on Age")
plt.xlabel("Average income (1,000s)")
plt.ylabel("Age (years)")
plt.grid(True)
plt.show()
```
